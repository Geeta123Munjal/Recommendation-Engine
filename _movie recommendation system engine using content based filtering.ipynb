```python
#selecting the features
features=['genre','country','director','actors']  
```

```python
# now we are combining the features
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
def combine_features(row):
      return row['genre']+" "+row['country']+" "+row['director']+" "+row['actors'] #all should be in string
data['combine_features']=data.apply(combine_features,axis=1) #axis=1 means we need to combine it vertically
print(data['combine_features']) #combinning the features and getting document of n dimension known AS Vector
```

```python
def get_title_from_index(index):
    return data[data.filmtv_id==index]["title"].values[0]
def get_index_from_title(title):
    return data[data.title==title]["filmtv_id"].values[0]
```

```python

cv=CountVectorizer()
count_matrix=cv.fit_transform(data['combine_features']) 
print(count_matrix.toarray())
```

```python
cosine_sim=cosine_similarity(count_matrix)
print(cosine_sim)
```

```python

movie_user_likes="Diner"
movie_index=get_index_from_title(movie_user_likes)
similar_movies=list(enumerate(cosine_sim[movie_index]))
sorted_similar_movies=sorted(similar_movies,key=lambda x:x[1],reverse=True)
i=0
for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0])) 
      
```
