```python
import pandas as pd
from scipy import sparse
from sklearn.metrics.pairwise import cosine_similarity
movies=pd.read_csv(r"C:\Users\Dell\Desktop\movies.csv")
ratings=pd.read_csv(r"C:\Users\Dell\Desktop\ratings.csv")
ratings=pd.merge(movies,ratings).drop(['genres','timestamp'],axis=1)
```

```python
ratings=ratings.pivot_table(index=['userId'],columns=['title'],values='rating')
ratings
```

```python
#remove movies that have user less than 10 and fill null values with zeroes have user greater than 10
ratings=ratings.dropna(thresh=10,axis=1).fillna(0)
ratings
```

```python
def standardize(row): #if a user don't give any ratinag, it does not mean that he  will give 0 rating so thats why we will find mean ratings
    new_row=(row-row.mean())/(row.max()-row.min())
    return new_row
user_ratings=ratings.apply(standardize)
user_ratings
```

```python
item_similarity=cosine_similarity(user_ratings.T)# we did transpose because we need to find item to item similariry(according to datset)
item_similarity
```

```python
# simplifying the item_similarity to understand
item_similarity_df=pd.DataFrame(item_similarity,index=ratings.columns,columns=ratings.columns)
item_similarity_df
```

```python
def get_similar_movies(movies_name,user_rating):
    similar_score=item_similarity_df[movies_name]*user_rating
    similar_score=similar_score.sort_values(ascending=False)
    return similar_score
```

```python
user1like=[("12 Years a Slave (2013)",4),("xXx (2002)",3),("Arsenic and Old Lace (1944)",2 )]
similar_movies=pd.DataFrame()
for movie,rating in user1like:
    similar_movies=similar_movies.append(get_similar_movies(movie,rating),ignore_index=True)
similar_movies.head()
similar_movies.sum().sort_values(ascending=False)
```
